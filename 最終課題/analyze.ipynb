```python
df_year = pd.read_sql("""
SELECT
  time,
  COUNT(*) AS n,
  AVG(value) AS avg_value
FROM observations
WHERE value IS NOT NULL
GROUP BY time
ORDER BY time
""", conn)

df_year
```

```python
df_area = pd.read_sql("""
SELECT
  area,
  COUNT(*) AS n,
  AVG(value) AS avg_value
FROM observations
WHERE value IS NOT NULL
  AND area IS NOT NULL
  AND area <> ''
GROUP BY area
ORDER BY avg_value DESC
LIMIT 10
""", conn)

df_area
```

```python
plt.figure()
plt.bar(df_area["area"], df_area["avg_value"])
plt.xlabel("Area")
plt.ylabel("Average Value")
plt.title("Top Areas by Average Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
